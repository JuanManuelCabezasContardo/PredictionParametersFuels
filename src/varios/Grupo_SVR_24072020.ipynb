```python
import numpy as np
import matplotlib
import pandas as pd
from datetime import datetime
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import median_absolute_error
from sklearn.metrics import mean_squared_log_error
import joblib
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
```

```python
fueldata=pd.read_csv('PWR_UOX-MOX_BigDataFrame-SF-GSRC.csv')
```

```python
for columna in ['fuelType',"Unnamed: 0","reactorType"]:
    fueldata= fueldata.loc[:, fueldata.columns != columna] 
fueldata
```

```python

#Neodimios
Neodimios = []
for column in fueldata.columns:
    if "Nd" in column: 
        Neodimios.append(column)
    if "TOT_" in column:
        Neodimios.append(column)
print(Neodimios)
X=fueldata[Neodimios]
Y=fueldata[['BU', 'CT', 'IE']]
X_CT=fueldata[Neodimios]
print(len(Neodimios))
#index de 5 años
index = Y.loc[((5*365) <= Y.CT)].index
Y["CT_anios"] = Y["CT"]/365
Y["CT_dias"] = Y["CT"]

```

```python
#para tener el grupo referencia que dejo requerimiento
```

```python
from sklearn.preprocessing import StandardScaler
X_s = StandardScaler()
X_ct_s=StandardScaler()

X_escalada=  X_s.fit_transform(X.values)
X_CT_escalda = X_ct_s.fit_transform(X_CT.iloc[index].values)
#Neodimios (Nd) junto con los Tot SF, GSRC y A,
```

```python
_conjunto_de_train_X,_conjunto_de_test_X,_conjunto_de_train_Y,_conjunto_de_test_Y=train_test_split(X_escalada,Y,test_size=0.2,random_state=128)#conjunto_X_de20000,conjunto_Y_de20000
_conjunto_CT_de_train_X,_conjunto_CT_de_test_X,_conjunto_CT_de_train_Y,_conjunto_CT_de_test_Y=train_test_split(X_CT_escalda,Y.iloc[index],test_size=0.2,random_state=128)#conjunto_X_de20000,conjunto_Y_de20000
print(_conjunto_de_train_X.shape,_conjunto_de_test_X.shape)
print(_conjunto_de_train_Y.shape,_conjunto_de_test_Y.shape)
print(_conjunto_CT_de_train_X.shape,_conjunto_CT_de_test_X.shape)
print(_conjunto_CT_de_train_Y.shape,_conjunto_CT_de_test_Y.shape)
```

```python
MODELO_SVR_BU= SVR(max_iter=1000)
MODELO_SVR_IE= SVR(max_iter=1000)
MODELO_SVR_CT_dias= SVR(max_iter=1000)
MODELO_SVR_CT_anios= SVR(max_iter=1000)
```

```python
print(datetime.now(),"BU")
MODELO_SVR_BU.fit(_conjunto_de_train_X,_conjunto_de_train_Y["BU"].values)
print(datetime.now(),"IE")
MODELO_SVR_IE.fit(_conjunto_de_train_X,_conjunto_de_train_Y["IE"].values)
print(datetime.now(),"CT")
MODELO_SVR_CT_dias.fit(_conjunto_CT_de_train_X,_conjunto_CT_de_train_Y["CT_dias"].values)
print(datetime.now(),"fin")
MODELO_SVR_CT_anios.fit(_conjunto_CT_de_train_X,_conjunto_CT_de_train_Y["CT_anios"].values)
print(datetime.now(),"fin")
```

```python
from joblib import dump
from joblib import load
MODELO_SVR_BU = joblib.load("SVR_BU_Grupo_2407.model")
MODELO_SVR_IE = joblib.load("SVR_IE_Grupo_2407.model")
MODELO_SVR_CT_dias = joblib.load("SVR_CT_dias_Grupo1_2407.model")
MODELO_SVR_CT_anios = joblib.load("SVR_CT_anios_Grupo_2407.model")
```

```python
print(datetime.now(),"BU")
Vector_entrenamiento_BU=MODELO_SVR_BU.predict(_conjunto_de_train_X)
print(datetime.now(),"IE")
Vector_entrenamiento_IE=MODELO_SVR_IE.predict(_conjunto_de_train_X)
print(datetime.now(),"CT")
Vector_entrenamiento_CT_dias=MODELO_SVR_CT_dias.predict(_conjunto_CT_de_train_X)
print(datetime.now(),"fin")
Vector_entrenamiento_CT_anios=MODELO_SVR_CT_anios.predict(_conjunto_CT_de_train_X)
print(datetime.now(),"fin")
```
