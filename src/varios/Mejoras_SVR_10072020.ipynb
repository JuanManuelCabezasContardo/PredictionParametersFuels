```python
import numpy as np
import matplotlib
import pandas as pd
from datetime import datetime
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import median_absolute_error
from sklearn.metrics import mean_squared_log_error
import joblib
from sklearn.svm import SVR
```

```python
fueldata=pd.read_csv('data_escalada_por_maximo.csv')
```

```python
for columna in ['fuelType',"Unnamed: 0","reactorType"]:
    fueldata= fueldata.loc[:, fueldata.columns != columna] 
fueldata
X=fueldata
Y=fueldata[['BU', 'CT', 'IE']]
for columna in ['BU', 'CT', 'IE']:
    X= X.loc[:, X.columns != columna] 
    X = X.loc[:, X.columns != columna]

from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(X.values,Y.values)
X_PCA= pca.transform(X)
X_PCA= pd.DataFrame(data = X_PCA, columns=["PCA_1","PCA_2","PCA_3"])
X_PCA
```

```python
_conjunto_de_train_X,_conjunto_de_test_X,_conjunto_de_train_Y,_conjunto_de_test_Y,=train_test_split(X,Y,test_size=0.20,random_state=42)
```

```python
print(_conjunto_de_train_X.shape,_conjunto_de_test_X.shape)
print(_conjunto_de_train_Y.shape,_conjunto_de_test_Y.shape)
```

```python
MODELO_SVR_BU= SVR(max_iter=100)
```

```python
"""
PCA_data =X_PCA
PCA_data["BU"]=Y["BU"]
PCA_data["CT"]=Y["CT"]
PCA_data["IE"]=Y["IE"]
PCA_data.to_csv("PCA_data.csv", index = False, header=True)
"""
```

```python
print(datetime.now())
MODELO_SVR_BU.fit(_conjunto_de_train_X.values,_conjunto_de_train_Y["BU"].values)
print(datetime.now())
```

```python
Vector_entrenamiento_BU=MODELO_SVR_BU.predict(_conjunto_de_train_X.values)
```

```python
data_metrica = []
#BU
from sklearn.metrics import max_error
data_metrica.append(["BU","Max error",max_error(_conjunto_de_train_Y["BU"].values, Vector_entrenamiento_BU)])
from sklearn.metrics import mean_absolute_error
data_metrica.append(["BU","MeanAE",mean_absolute_error(_conjunto_de_train_Y["BU"].values, Vector_entrenamiento_BU)])
from sklearn.metrics import mean_squared_error
data_metrica.append(["BU","MSE",mean_squared_error(_conjunto_de_train_Y["BU"].values, Vector_entrenamiento_BU)])
from sklearn.metrics import mean_squared_log_error
data_metrica.append(["BU","MSLE",mean_squared_log_error(_conjunto_de_train_Y["BU"].values, Vector_entrenamiento_BU)])
from sklearn.metrics import median_absolute_error
data_metrica.append(["BU","MedianAE",median_absolute_error(_conjunto_de_train_Y["BU"].values, Vector_entrenamiento_BU)])
from sklearn.metrics import r2_score
data_metrica.append(["BU","R2",r2_score(_conjunto_de_train_Y["BU"].values, Vector_entrenamiento_BU)])
```

```python
data_metrica = pd.DataFrame(data=data_metrica, columns=["Prediccion","metrica","Con Batch"])
data_metrica.style.format({'Con Batch':"{:+.3f}"})
```
